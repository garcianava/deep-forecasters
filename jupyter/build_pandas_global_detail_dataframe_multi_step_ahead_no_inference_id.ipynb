```python
model_id = 'DMSLSTM_TPU_003'
execution = 0
dataset = 'test'
```

```python
flag = \
df.model_id.eq(model_id) & \
df.execution.eq(execution) & \
df.dataset.eq(dataset)
```

```python
df[flag]
```

```python
plots = dict()
```

```python
# a colors list with 12 different color, repeated once
colors_list = d3['Category20'][12] + d3['Category20'][12]
```

```python
exec_label = 'execution_{}'.format(execution)
```

```python
plots[exec_label] = figure(
    x_axis_type='datetime',
    plot_width=960,
    plot_height=400,
    title='Multi-step prediction for model {}, execution {}, on {} dataset'.format(model_id,
                                execution,
                                dataset))

plots[exec_label].grid.grid_line_alpha=0.5

plots[exec_label].xaxis.axis_label = 'Date'
plots[exec_label].yaxis.axis_label = 'Value'

for n in [0, 3, 7, 11, 15, 19, 23]:
    timestamps = [pd.to_datetime(row.string_timestamps[n]) for _, row in df[flag].iterrows()]
    targets = [row.targets[n] for _, row in df[flag].iterrows()]
    predictions = [row.predictions[n] for _, row in df[flag].iterrows()]
    
    plots[exec_label].line(
        timestamps,
        targets,
        color='#A6CEE3',
        legend_label='real')

    plots[exec_label].line(
            timestamps,
            predictions,
            color=colors_list[n],
            legend_label='{}-th timestep'.format(n))

# uncomment the following two lines to save plot
# output_file('/home/developer/gcp/cbidmltsf/datasets/cfe/{}_H_kw.html'.format(device))
# save(fig_kw)

# uncomment the following line to display plot
show(plots[exec_label])
```
