```python
def process_list(s, variable_type=int, delimiter=None):
    """Parses a line in the PEMS format to a list."""
    if delimiter is None:
      l = [
          variable_type(i) for i in s.replace('[', '').replace(']', '').split()
      ]
    else:
      l = [
          variable_type(i)
          for i in s.replace('[', '').replace(']', '').split(delimiter)
      ]

    return l
```

```python
def read_single_list(filename):
    """Returns single list from a file in the PEMS-custom format."""
    with open(os.path.join(data_folder, filename), 'r') as dat:
        l = process_list(dat.readlines()[0])
    return l
```

```python
def get_learning_rate_vaswani(params, current_step):
    # from Vaswani et al., 2017
    learning_rate = params['d_model']**params['exp_1'] * min(
        current_step**params['exp_2'], current_step*params['warmup_steps']**(params['exp_2'] - 1)
    )
    # modify according to batch_size
    learning_rate *= (256./params['train_batch_size'])
    return learning_rate
```

```python
params = {
    'd_model': 256,
    'train_batch_size': 256,
    'train_steps': 15000,
    'warmup_steps': 7500,
    'exp_1': -0.5,
    'exp_2': -0.5
}
```

```python
steps = 1 + np.arange(params['train_steps'])
steps
```

```python
first_part = [params['d_model']**params['exp_1']*step**params['exp_2'] for step in steps]
```

```python
second_part = [params['d_model']**params['exp_1']*step*params['warmup_steps']**(params['exp_2'] - 1) for step in steps]
```

```python
learning_rates = np.array([get_learning_rate_vaswani(params, step) for step in steps])
```
