```python
import numpy as np
import pandas as pd
```

```python
from bokeh.plotting import figure, show, output_file, save
from bokeh.io import output_notebook
from bokeh.layouts import column
from bokeh.palettes import d3
output_notebook()
```

```python
pickle_path = '/home/developer/gcp/cbidmltsf/sldbs/LD2011-2014_SEPARATED_FULL_BSCTRFM_168_168_07DB_MMX/test'
```

```python
ts_df = pd.read_pickle('{}/MT_320.pkl'.format(pickle_path))
ts_df
```

```python
plots = dict()
```

```python
plots['data'] = figure(
    plot_width=960,
    plot_height=320,
    tools='',
    x_axis_type='datetime'
)

plots['data'].line(
    x=ts_df.date,
    y=ts_df.kw_scaled,
    color='black'
)

plots['sin_hour_day'] = figure(
    plot_width=960,
    plot_height=320,
    tools='',
    x_axis_type='datetime'
)

plots['sin_hour_day'].line(
    x=ts_df.date,
    y=ts_df.sin_hour_day,
    color='black'
)

plots['sin_day_week'] = figure(
    plot_width=960,
    plot_height=320,
    tools='',
    x_axis_type='datetime'
)

plots['sin_day_week'].line(
    x=ts_df.date,
    y=ts_df.sin_day_week,
    color='black'
)

show(column(plots['data'], plots['sin_hour_day'], plots['sin_day_week']))
```
