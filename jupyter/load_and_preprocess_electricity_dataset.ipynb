```python
import pandas as pd
```

```python
from bokeh.plotting import figure, show, output_file, save
from bokeh.io import output_notebook
# select a palette
from bokeh.palettes import d3
output_notebook()
```

```python
DATASET_DIR = '/home/developer/gcp/cbidmltsf/datasets/electricity/LD2011_2014.txt'
```

```python
dataset_filename = '{}/LD2011_2014.txt'.format(DATASET_DIR)
```

```python
# a file input to iterate on
file_input = open(dataset_filename, 'rt')
```

```python
# a text file to store the dataset, decimal commas replaced with decimal points
filename_processed = '{}/LD2011_2014_dots.txt'.format(DATASET_DIR)
filename_processed
```

```python
# a file output to iterate on
file_output = open(filename_processed, 'wt')
```

```python
for line in file_input:
    file_output.write(line.replace(',', '.'))

file_input.close()
file_output.close()
```

```python
# load the dataset from the file with decimal points
df = pd.read_csv(filename_processed, sep=";", index_col=0)
```

```python
# convert the dataframe index from string to timestamp
df.index = pd.to_datetime(df.index)
```

```python
# now dataframe can be persisted as Pandas pickle, no pre-processing yet
raw_pickle_filename = '{}/raw.pkl'.format(DATASET_DIR)
raw_pickle_filename
```

```python
df.to_pickle(raw_pickle_filename)
```

```python
# a dictionary to manage plots
plots = dict()
```

```python
customer_id = 'MT_320'

plots['customer_id'] = figure(
    title='Electric Load Diagram for {}'.format(customer_id),
    plot_width=960,
    plot_height=320,
    x_axis_type ='datetime',
    x_axis_label='Timestamp',
    y_axis_label='Electric Load')
plots['customer_id'].line(x=df['MT_001'].index,
                          y=df['MT_001'].values,
                          legend_label=customer_id,
                          color='blue',
                          line_width=1)
plots['customer_id'].legend.location = 'top_center'

show(plots['customer_id'])
```
