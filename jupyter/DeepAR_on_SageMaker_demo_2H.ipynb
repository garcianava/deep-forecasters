```python
filepath = '{}/data/train/train.json'.format(PROJECT_ROOT)
```

```python
# the json file is a collection of dictionaries {"start": timestamp, "target": time-series}
# each row in the json file is a different dictionary
```

```python
# open the collection of dictionaries
file = open(filepath, 'r')

# and get an iterable line by line (dict by dict)
all_lines = file.readlines()
```

```python
train = dict()
```

```python
counter = 1

for line in all_lines:
    customer_id = 'MT_{:03d}'.format(counter)
    train[customer_id] = json.loads(line)
    counter+=1
```

```python
train.keys()
```

```python
for key in train.keys():
    print('Series {} starts at {} and has {} values...'.format(
        key,
        train[key]['start'],
        len(train[key]['target'])
    )
)
```
